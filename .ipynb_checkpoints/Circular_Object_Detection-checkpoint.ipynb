## Circular Object Detection

### In an Image

```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
from pylab import rcParams            #Used for Enlarge the image with the same pixel size
rcParams['figure.figsize']=(5,5)    #Set the enlargement size
```

```python
#Import Original Image from the same folder of the file
circular_img=cv.imread("detect_circles.jpg")
plt.imshow(circular_img)  #It Shows image in BGR format
```

```python
# Convert it to RGB Format
circular_rgb=cv.cvtColor(circular_img,cv.COLOR_BGR2RGB)
output=circular_rgb.copy()
plt.imshow(circular_rgb)
```

```python
# Convert it to the GRAY scale Format  Because OpenCV will always work on GRAY Scale Format
circular_gray=cv.cvtColor(circular_rgb,cv.COLOR_RGB2GRAY)
plt.imshow(circular_gray)
```

# Detetcing the circles

```python
circles= cv.HoughCircles(circular_gray,cv.HOUGH_GRADIENT,1.2,220)
circles.shape
```

```python
circles
```

```python
# Ensures some circle were found
if circles is not None:
    #Convert the (x,y) coordinates & radius of circle to integers
    circles=np.round(circles[0,:]).astype('int')
    
    #Loop over (x,y) coordinate & radius of circle
    for (x,y,r) in circles:
        # Draw the circle in output image then draw the rectangle
        # corresponding to the center of the circle
        cv.circle(output,(x,y),r,[255,0,0],10)
        cv.rectangle(output,(x-5,y-5),(x+5,y+5),[255,0,0],-1)
        #Now we put the original image with this circle detected image & show them
        com=np.hstack([circular_rgb,output])
        plt.imshow(com)
       
        #Show the output image
        cv.imshow("Output",np.hstack([circular_rgb, output]))
        cv.waitKey(0)
```

## Complete Circular Object Detection In One Function for above image

```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
from pylab import rcParams            #Used for Enlarge the image with the same pixel size
rcParams['figure.figsize']=(5,5)    #Set the enlargement size
```

```python
def Circular_Object_Detection():
    
    #Import Original Image from the same folder of the file
    circular_img=cv.imread("detect_circles.jpg")
    
    # Convert it to RGB Format
    circular_rgb=cv.cvtColor(circular_img,cv.COLOR_BGR2RGB)
    output=circular_rgb.copy()
    
    # Convert it to the GRAY scale Format  Because OpenCV will always work on GRAY Scale Format
    circular_gray=cv.cvtColor(circular_rgb,cv.COLOR_RGB2GRAY)

    #Detetcing the circles
    circles= cv.HoughCircles(circular_gray,cv.HOUGH_GRADIENT,1.2,220)
    circles.shape

    circles

    # Ensures some circle were found
    if circles is not None:
        #Convert the (x,y) coordinates & radius of circle to integers
        circles=np.round(circles[0,:]).astype('int')
    
        #Loop over (x,y) coordinate & radius of circle
        for (x,y,r) in circles:
            # Draw the circle in output image then draw the rectangle
            # corresponding to the center of the circle
            cv.circle(output,(x,y),r,[255,0,0],10)
            cv.rectangle(output,(x-5,y-5),(x+5,y+5),[255,0,0],-1)
            #Now we put the original image with this circle detected image & show them
            com=np.hstack([circular_rgb,output])
            #plt.imshow(com)
       
        #Show the output image
        cv.imshow("Output",np.hstack([circular_rgb, output]))
        cv.waitKey(0)
```

```python
Circular_Object_Detection()
```
